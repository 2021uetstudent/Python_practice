**Palindrome Checker**

```python
str_variable = input("Enter one word to check palindrome ")
valid_str = str_variable.lower()
if(valid_str[0:] == valid_str[-1::-1]):
    print(f"{valid_str} is palindrome.")
else:
    print(f"{valid_str} is not plaindrome.")
```

**Words and Character Frequency Counter**

```python
from collections import Counter
# Input the string from the user
text = input("")
lower_text = text.lower()
split_text = lower_text.split()
letter = [ ]
for ch in split_text:
    if (ch.isalpha()):
        letter.append(ch)
        word_frequecy = Counter(ch)
        print(word_frequecy)


letter_frequency = Counter(letter)
print(letter_frequency)
```

**Vowel & Consonant Counter**

```python
vowel = ['a','e','i','o','u']
input_str = input()
v, con = 0,0
for ch in input_str:
    if ch in vowel:
        print(f"{ch} is vowel.")
        v= v+1  
    else:
        print(f"{ch} is consonant.")
        con= con +1
        
print(f"Total vowel character is {v} in {input_str}")
print(f"Total consonant in {input_str} is {con}")
```

**Replace all vowel with #** 

```python
# Loop method
text = input()
vowel= ['a','e','i','o','u','A','E','I','O','U']
#split_text = text.split()
for ch in text:
    if (ch in vowel):
        new_text = text.replace(ch,"#")
print(new_text)

# without Loop method
no_vowel_text = (text.replace("a","#").replace("e","#").replace("i","#")
.replace("o","#").replace("u","#").replace("A","#").replace("E","#")
.replace("I","#").replace("O","#").replace("U","#"))
print(no_vowel_text)
```

**Reverse the word order**

```python
input_string = input()
split_string = input_string.split()
reverse_words = split_string[::-1]
reverse_string = " ".join(reverse_words)
print(reverse_string)
```

**Anagram Checker**

```python
word1 = input()
word2 = input()
sorted_word1= sorted(word1)
sorted_word2= sorted(word2)
if (Counter(sorted_word1) == Counter(sorted_word2)):
    print(f"{word1} and {word2} is Anagram.")
else:
    print(f"{word1} and {word2} is not Anagram")
        
```
