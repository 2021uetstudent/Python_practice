**Find the index of the highest price ; 
Count how many prices are above 200**

```python
prices = [250, 89, 340, 15, 780, 99, 450]
highest_price = max(prices)
print(f"The index of the highest price: ", prices.index(highest_price))
count = 0
for price in prices:
    if price > 200:
        count = count + 1
    else:
        continue
print(f"The prices which are above 200: {count}")
```

**Create a new list with all words converted to uppercase.
Create a new list with only words longer than 3 characters.
Create a new list containing the length of each word.
Create a list of tuples: (word, length) for each word.
Replace each word with "short" if length ≤ 3, else "long".**

```python
words = ["data", "AI", "python", "sql", "excel", "ml"]
upper_letter_list = [ ]
long_char_list = [ ]
word_length_list = [ ]
tuple_list = [ ]
short_long_list = [ ]
for word in words:
    upper_letter_list.append(word.upper())
    word_length_list.append(len(word))
    tuple_list.append((word,len(word)))
    if len(word) > 3:
        long_char_list.append(word)
        short_long_list.append("long")
    else:
        short_long_list.append("short")
        continue
              
print(f"Upper letter list: {upper_letter_list}")
print(f"Words longer than 3 characters: {long_char_list}")
print(f"Length of each word: {word_length_list}")
print(f"Tuple list (word, length): {tuple_list}")
print(f"Short/Long classification: {short_long_list}")
```

**Create a list flagging each temp as "fever" if > 100, else "normal".
Get only the fever temperatures using comprehension**

```python
temps = [98, 102, 99, 105, 97, 100, 103]
flagging_list = ["Fever" if temp > 100 else "Normal" for temp in temps]
print(flagging_list)
fever_list = [temp for temp in temps if temp > 100 ]
print(fever_list)
```

**Nested Lists (Mini-Dataset Practice)**

```python
employees = [
    ["Ali", 28, "Sales", 50000],
    ["Sara", 34, "IT", 75000],
    ["John", 25, "Sales", 48000],
    ["Amna", 40, "HR", 60000],
    ["Bilal", 30, "IT", 82000]
]
# columns: Name, Age, Department, Salary
# Extract a list of all names
name_list = [row[0] for row in employees]
print(f"Name list of employees:{name_list}")
# Extract a list of all salaries
salary_list= [row[3] for row in employees]
print(f"Salary list of employees:{salary_list}")
#Find the average salary
Average_salary = (sum(salary_list) / len(salary_list))
print(f"Average salary: {Average_salary}")
# Find the employee (full row) with the highest salary
highest_salary = max(salary_list)
print(highest_salary)
highest_salary_employee = [employees [j] for j in range(0,(len(employees))) if (employees[j][3] == highest_salary)]
print(f"Highest salary employee: {highest_salary_employee}")
#Get all employees who work in "IT"
IT_employee = [employees [k] for k in range(0,(len(employees))) if (employees[k][2] == "IT")]
print(f"Employees who work in IT: {IT_employee}")
# Get names of employees older than 30.
senior_name_list = [employees[i][0] for i in range(0,(len(employees))) if employees[i][1] > 30]
print(f"Name list of employees:{senior_name_list}")
# Calculate total salary paid to the "Sales" department
paid_salary =[row[3] for row in employees if row[2] == "Sales"]
total_paid_salary = sum(paid_salary)
print(f"total salary paid to the Sales department: {total_paid_salary}")
# Create a list of [name, salary] pairs only.
name_salary_pair_list =[[row[0],row[3]] for row in employees]
print(f"Name Salary pair list: {name_salary_pair_list}")
# unique Department in the Dataset
unique_department_list =set([employees[k][2] for k in range(len(employees))])
print(unique_department_list)
# Sort the employees list by salary
sorted_employee_list = sorted(employees, key= lambda row : row[3])
print(f"Sorted employee list by salary: {sorted_employee_list}")
```

```python
# Find the longest streak of consecutive non-zero sales days
daily_sales = [120, 340, 0, 450, 230, 0, 670, 890, 0, 120]
max_streak, counter_streak = 0, 0

for value in daily_sales:
    if value != 0:                     # non-zero check
        counter_streak += 1
        max_streak = max(max_streak, counter_streak)
    else:
        counter_streak = 0              # reset on zero (streak break)

print(f"Longest consecutive non-zero streak: {max_streak}")
```

```python
# Create a running total (cumulative sum) list 
daily_sales = [120, 340, 0, 450, 230, 0, 670, 890, 0, 120]
cumulative_sum = []
running_total =0
for i in range(len(daily_sales)):
    running_total = daily_sales[i] + running_total
    cumulative_sum.append(running_total)

print("Cumulative sum of the list:",cumulative_sum)
```
