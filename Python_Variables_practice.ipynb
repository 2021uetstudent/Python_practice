**Record of Students who enrolled in the Data Analyst Course**
Store your name, education, University, and course  in variables, 
then print a formatted bio.

```python
number_of_students = int(input("How many student's record to print? "))
for i in range(number_of_students):
    name = input("Student Name: ")
    education = input("Last degree: ")
    education_institute = input("Enter the Institute name from where you have completed your last degree: ")
    course = input("Course to enroll: ")
    if course.lower() != "data analyst":
      print("please enter the valid course (Data Analyst)")
  
    print(f"Student{i+1} Record")
    print(f"Student Name:{name}\nQualification:{education}\nUniversity:{education_institute}\nCourse:{course}\n\n")
```

**Circle Stats**
Store a radius, then calculate the area and circumference using pi = 3.14.

```python
import math

try:
   PI = 3.14
   radius = int(input("Enter the radius to calculate circle stats: "))
   if type(radius) != str:
    area =  PI * radius**2 
    circumference = 2* PI * radius
    print(f"Area of Circle ={area}\nCircumference of Circle= {circumference}\n\n")
    
except ValueError:
  print("Please enter the valid radius of circle")
```

**Find the middle number**

```python
num1 = int(input())
num2 = int(input())
num3 = int(input())
print(f"Numbers are: {num1},{num2},{num3}")
if (num1 <= num2 <=num3) or (num3<= num2 <=num1):
  print(f"{num2} is middle number")
elif (num2<= num1 <= num3) or (num3<= num1 <= num2):
  print(f"{num1} is middle number")
else:
  print(f"{num3} is middle number")
```

**Digit Extraction Game**

```python
number = int(input("Enter a number: "))
print(f"Number to extract is {number}")
digit = len(str(number))

while(digit > 0):
    output = number // (10 ** (digit - 1))
    print(output)
    number = number % (10 ** (digit - 1))
    digit -= 1
```
